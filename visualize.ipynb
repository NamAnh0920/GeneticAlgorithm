```python
import seaborn as sns
import pandas as pd

url = './summary_25_50_100_duplicate.csv'

data = pd.read_csv(url)
data
```

```python
def draw_heatmap(*args, **kwargs):
    data = kwargs.pop('data')
    d = data.pivot(index=args[0], columns=args[1], values=args[2])
    sns.heatmap(d, **kwargs)

def visualize(data, value):
    min_val = min(data[value])
    max_val = max(data[value])
    g = sns.FacetGrid(data, row = 'population', col = 'length', aspect=1)
    cbar_ax = g.fig.add_axes([.92, .3, .02, .4])
    g.map_dataframe(draw_heatmap, 'n_mutation', 'n_crossover', value, vmin = min_val, vmax = max_val, cbar_ax = cbar_ax)
    g.tight_layout(rect=[0, 0, .9, 1])
    g.savefig(f'./visualize/{value}_25.jpg')

def visualize_data(data, target):
    for value in target:
        visualize(data, value)

visualize_data(data, ['epoch', 'time', 'max_score'])
```
